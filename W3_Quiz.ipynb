```python
import graphlab
```

```python
products = graphlab.SFrame('amazon_baby.gl/.')
```

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

# Build the word count vector for each review

```python
products['word_count'] = graphlab.text_analytics.count_words(products['review'])
```

```python
products.head()
```

### wirte a for loop although it was not recommand

```python
for i in range(len(products['word_count'])):
    if 'and' in products['word_count'][i]:
        continue
    else:
        products['word_count'][i]['and'] = 0
products['and'] = products['word_count'].apply(lambda x: x['and'])
```

### use the .apply() to do the job as recommand

```python
def count_and(count):
    if 'and' in count:
        return count['and']
    else:
        return 0
```

```python
products['and'] = products['word_count'].apply(count_and)
```

## build a loop to call all the name

```python
del products['and']
```

```python
for i in range(len(selected_words)):
    selected_words[i]
    def count_words(words):
        if selected_words[i] in words:
            return words[selected_words[i]]
        else:
            return 0
    products[selected_words[i]] = products['word_count'].apply(count_words)    
```

## Build a sentimental analysis model

### Define the sentiment

```python
products['rating'].show(view='Categorical')
```

```python
products = products[products['rating']!=3]
```

```python
def cal_rate(rate):
    if rate >= 4:
        return 1
    else:
        return 0
```

```python
products['sentimenty'] = products['rating'].apply(cal_rate)
```

```python
products.head()
```

```python
train_data,test_data = products.random_split(.8, seed=0)
```

```python
selected_words_model = graphlab.logistic_classifier.create(train_data,
                                                     target = 'sentimenty',
                                                     features = selected_words,
                                                     validation_set = test_data)
```

```python
selected_words_model['coefficients']
```

# Evaluate the sentiment model

```python
selected_words_model.evaluate(test_data)
```

```python
diaper_champ_reviews = products[products['name'] == 'Baby Trend Diaper Champ']
```

# Apply the learned model to understand sentiment for diaper_champ

```python
diaper_champ_reviews['predict_sentiment'] = selected_words_model.predict(diaper_champ_reviews,output_type='probability')
diaper_champ_reviews.head()
```

## Sort the review based on the predicted sentiment and explore 

```python
diaper_champ_reviews = diaper_champ_reviews.sort('predict_sentiment',ascending=False)
```
